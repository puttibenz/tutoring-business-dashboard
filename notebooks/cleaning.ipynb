```python
# 2. ฟังก์ชันอัจฉริยะสำหรับจัดการชื่อติวเตอร์
def clean_tutor_names(tutor_str):
    if pd.isna(tutor_str):
        return tutor_str
        
    # แยกชื่อติวเตอร์ด้วยเครื่องหมายลูกน้ำ
    names = [name.strip() for name in str(tutor_str).split(',')]
    
    unique_names = []
    seen_cores = set() # เอาไว้จดจำว่าใครถูกดึงชื่อไปแล้วบ้าง
    
    for name in names:
        if not name: 
            continue
            
        # เคลียร์เว้นวรรคที่ผิดปกติ (เช่น "( อ. แชร์ )" -> "(อ.แชร์)")
        clean_name = re.sub(r'\s+', ' ', name)
        clean_name = clean_name.replace("( ", "(").replace(" )", ")")
        clean_name = clean_name.replace("อ. ", "อ.")
        
        # สร้าง Core Name สำหรับเทียบตัวซ้ำ (ลบช่องว่างทั้งหมด และตัดคำว่า "ดร." ออก)
        # ตัวอย่าง: "ดร.ธรรมนิติ์ พิพัฒน์ศรีสวัสดิ์ (อ.นอต)" จะเหลือแค่ "ธรรมนิติ์พิพัฒน์ศรีสวัสดิ์(อ.นอต)"
        core_name = clean_name.replace(" ", "")
        core_name = re.sub(r'^(ดร\.)', '', core_name)
        
        # ถ้าชื่อแกนหลัก (Core Name) นี้ยังไม่เคยมีในลิสต์ ให้เพิ่มเข้าไป
        if core_name not in seen_cores:
            seen_cores.add(core_name)
            unique_names.append(clean_name)
            
    # นำชื่อที่ไม่ซ้ำแล้วมาต่อกันด้วยลูกน้ำเหมือนเดิม
    return ", ".join(unique_names)

# 3. ลองพริ้นต์ดูความเปลี่ยนแปลงก่อน (Before)
print("--- ก่อน Clean ---")
print(df[df['tutor'].str.contains('ดร.') | df['tutor'].str.contains('อ. แชร์')]['tutor'].head(3).values)

# 4. นำฟังก์ชันไปประยุกต์ใช้กับคอลัมน์ tutor ทัังหมด
df['tutor'] = df['tutor'].apply(clean_tutor_names)

print("\n--- หลัง Clean ---")
print(df[df['tutor'].str.contains('ธรรมนิติ์') | df['tutor'].str.contains('อ.แชร์')]['tutor'].head(3).values)
```

```python
# 2. ฟังก์ชันอัจฉริยะสำหรับจัดหมวดหมู่วิชาจากชื่อคอร์ส
def clean_vague_subjects(row):
    name = str(row['course_name']).lower()
    old_subject = str(row['subject'])
    
    # --- ขั้นที่ 1: ตรวจจับ Keyword สำคัญในชื่อคอร์ส ---
    # นับว่ามีวิชาสายวิทย์กี่วิชา
    sci_count = sum(1 for sub in ['ฟิสิกส์', 'เคมี', 'ชีวะ', 'phy', 'chem', 'bio'] if sub in name)
    
    has_math = any(sub in name for sub in ['คณิต', 'math'])
    has_thai = 'ไทย' in name
    has_soc = 'สังคม' in name
    has_eng = any(sub in name for sub in ['อังกฤษ', 'eng'])
    
    # ตรวจจับข้อสอบเฉพาะทาง
    has_med = any(sub in name for sub in ['tpat1', 'กสพท', 'แพทย์'])
    has_eng_tech = any(sub in name for sub in ['tpat3', 'วิศวะ', 'ความถนัดวิศวะ'])
    has_aptitude = any(sub in name for sub in ['tgat', 'tpat2', 'tpat4', 'tpat5', 'ความถนัด'])

    # --- ขั้นที่ 2: จัดกลุ่มหมวดหมู่ใหม่ (เรียงจากแพ็กเกจใหญ่สุด ไปหาวิชาเดี่ยว) ---
    
    # 2.1 กลุ่มแพ็กเกจสอบเข้าเฉพาะทาง (มักจะรวมหลายวิชา)
    if has_med and (has_math or sci_count > 0 or has_eng):
        return 'แพ็กเกจเตรียมสอบแพทย์ (กสพท/TPAT1)'
    if has_eng_tech and (has_math or sci_count > 0):
        return 'แพ็กเกจเตรียมสอบวิศวะ (TPAT3)'
        
    # 2.2 กลุ่มแพ็กเกจรวมวิชาพื้นฐาน
    if sci_count >= 2 and has_math:
        return 'แพ็กเกจวิทย์-คณิต (รวมวิชา)'
    if sci_count >= 2 and not has_math:
        return 'วิทยาศาสตร์รวม (ฟิสิกส์/เคมี/ชีวะ)'
    if has_thai and has_soc:
        return 'ภาษาไทย-สังคมศึกษา'
        
    # 2.3 กลุ่มสอบความถนัด (TGAT/TPAT ล้วนๆ ที่ไม่ได้ผูกกับวิชาสามัญเยอะๆ)
    if has_aptitude or has_med or has_eng_tech:
        return 'ความถนัด (TGAT/TPAT/สอบเฉพาะทาง)'

    # 2.4 กลุ่มวิชารายตัว (Single Subjects)
    if has_math and sci_count == 0: return 'คณิตศาสตร์'
    if 'ฟิสิกส์' in name or 'phy' in name: return 'ฟิสิกส์'
    if 'เคมี' in name or 'chem' in name: return 'เคมี'
    if 'ชีวะ' in name or 'bio' in name: return 'ชีววิทยา'
    if has_eng: return 'ภาษาอังกฤษ'
    if has_thai and not has_soc: return 'ภาษาไทย'
    if has_soc and not has_thai: return 'สังคมศึกษา'

    # --- ขั้นที่ 3: ถ้ากวาดหมดแล้วยังไม่เจอ ให้คืนค่าเดิม แต่คลีนชื่อให้สวยขึ้น ---
    if old_subject == 'Mixed':
        return 'แพ็กเกจรวมอื่นๆ'
        
    return old_subject

# 3. ลองเช็คข้อมูลก่อน Clean
print("--- สัดส่วนหมวดหมู่วิชา ก่อน Clean ---")
print(df['subject'].value_counts().head(10))

# 4. เรียกใช้งานฟังก์ชัน (สังเกตว่าเราใช้ axis=1 เพื่อส่งข้อมูลไปทั้งแถว)
df['subject'] = df.apply(clean_vague_subjects, axis=1)

print("\n--- สัดส่วนหมวดหมู่วิชา หลัง Clean ---")
print(df['subject'].value_counts().head(10))
```

```python
print(f"--- จำนวนคอร์สทั้งหมดก่อนลบตัวซ้ำ: {len(df)} คอร์ส ---")

# 2. (Option) แอบดูข้อมูลที่ซ้ำกันก่อนลบ เพื่อความชัวร์
# keep=False หมายถึง ให้โชว์คอร์สที่ซ้ำกันออกมาให้ดู "ทุกบรรทัด" (ทั้งตัวต้นฉบับและตัวก๊อปปี้)
duplicates = df[df.duplicated(subset=['institute_name', 'course_name'], keep=False)]

if not duplicates.empty:
    print("\n[!] พบข้อมูลคอร์สที่ชื่อซ้ำกันดังนี้:")
    # โชว์แค่ชื่อสถาบัน ชื่อคอร์ส และ URL เพื่อให้เห็นว่ามันต่างกันที่ URL จริงๆ
    print(duplicates[['institute_name', 'course_name', 'url']].sort_values(by='course_name'))
else:
    print("\n[!] ไม่พบข้อมูลซ้ำ")

# 3. จัดการลบข้อมูลซ้ำ (Drop Duplicates)
# subset=['institute_name', 'course_name'] -> ให้เช็คความซ้ำจาก 2 คอลัมน์นี้คู่กัน
# keep='first' -> ถ้าเจอซ้ำกัน ให้เก็บ URL อันแรกที่บอทกวาดเจอเอาไว้ แล้วลบอันที่เหลือทิ้ง
df_clean = df.drop_duplicates(subset=['institute_name', 'course_name'], keep='first')

print(f"\n--- จำนวนคอร์สทั้งหมดหลังลบตัวซ้ำ: {len(df_clean)} คอร์ส ---")
```

```python
full_course_keywords = [
    'a-level', 'tcas', 'tgat', 'tpat', 
    'upscore', 'ติดอาวุธลับ', 'แพ็กเกจเตรียมสอบแพทย์', 
    'แพ็กเกจเตรียมสอบวิศวะ','ม.4','ม.5','ม.6',
    'เทอม', 'entrace', 'สอบเข้า', 'แพ็กเกจรวม', 'รวมวิชา', 
    'แพ็กเกจเตรียมสอบ', 'แพ็กเกจวิทย์-คณิต','pack',
    'เล่ม', 'v.เต็ม'
]

is_full_by_name = df['course_name'].str.contains('|'.join(full_course_keywords), case=False, na=False)

is_full_by_hours = df['total_hours'] >= 30
is_package_type = df['course_type'] == 'คอร์สแพ็กเกจ'

df['course_scope'] = np.where(is_full_by_name | is_full_by_hours | is_package_type, 
                              'Full Course (คอร์สเตรียมสอบ/รวมเทอม)',
                              'Chapter (คอร์สเนื้อหาแยกบท)')

print(df[['course_name', 'course_scope']].head(10))
```

```python
# ── ตรวจสอบ: ตัวอย่าง Chapter (ควรเป็นคอร์สแยกบทจริงๆ) ──────────────────
print("=== ตัวอย่าง Chapter (แยกบท) — 15 รายการ ===")
print(
    df[df['course_scope'] == 'Chapter (คอร์สเนื้อหาแยกบท)']
    [['institute_name', 'course_name', 'total_hours', 'price', 'course_type']]
    .sort_values('institute_name')
    .head(15)
    .to_string(index=False)
)

print("\n=== ตัวอย่าง Full Course — 10 รายการ (hours < 30 แต่ถูก classify เป็น Full) ===")
edge = df[
    (df['course_scope'] == 'Full Course (คอร์สเตรียมสอบ/รวมเทอม)') &
    (df['total_hours'] < 30)
]
print(edge[['institute_name', 'course_name', 'total_hours', 'price', 'course_type']].head(10).to_string(index=False))
print(f"\nรวม edge-case (Full แต่ hours < 30): {len(edge)} รายการ")

print("\n=== สถิติ hours ของแต่ละ scope ===")
print(df.groupby('course_scope')['total_hours'].describe().round(1))
```

```python
# ── ตรวจ False Negative: หาคอร์สที่น่าจะเป็น Chapter แต่หลุดไปเป็น Full ──
# Pattern เช่น [บทที่ X] ควรเป็น Chapter เสมอ โดยไม่ขึ้นกับ hours
import re
hard_chapter_pattern = re.compile(r'\[บทที่\s*\d+\]', re.IGNORECASE)
false_neg = df[
    (df['course_scope'] == 'Full Course (คอร์สเตรียมสอบ/รวมเทอม)') &
    df['course_name'].str.contains(hard_chapter_pattern, regex=True)
]
print(f"False Negative (มี [บทที่ X] แต่ถูก classify เป็น Full): {len(false_neg)} รายการ")
if not false_neg.empty:
    print(false_neg[['course_name', 'total_hours', 'course_type']].to_string(index=False))

# ── ถ้า False Negative มี → แก้ด้วย hard rule ────────────────────────────
if not false_neg.empty:
    df.loc[df['course_name'].str.contains(hard_chapter_pattern, regex=True), 'course_scope'] = \
        'Chapter (คอร์สเนื้อหาแยกบท)'
    print("\n[✓] แก้ไข False Negative แล้ว")
    print(df['course_scope'].value_counts())
```
